```python
years=[2000, 2005, 2010, 2015]
```

```python
energy=['a', 'b', 'c', 'd']
```

```python
fig, axs=plt.subplots(2,2)
axs[0,0].plot(years, energy, '-b')
axs[0,0].set_title('a')
axs[0,1].plot(years, energy, 'tab:red')
axs[0,1].set_title('b')
axs[1,0].scatter(years, energy, c='yellow')
axs[1,0].set_title('c')
axs[1,1].scatter(years, energy, c='green')
axs[1,1].set_title('d')

```
